```python
results = {
    "EbN0": [0,2,4,6,8],
    "BER": [0.12,0.05,0.01,0.0008,0.0001],
    "iterations": 10000
}

import pickle
# wb - write binary
with open("results.pkl", "wb") as f:
    pickle.dump(results, f)
```

```python
with open("results.pkl", "rb") as f:
    result = pickle.load(f)
print(result)
```
