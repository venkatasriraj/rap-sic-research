```python
a = 10  ## 1010
b = 5   ## 0101
print(f"Bit-wise AND {a & b}")
print(f"Bit-wise OR {a | b}")
print(f"Bit-wise XOR {a ^ b}")
print(f"Bit-wise right shift {a >> 1}")
print(f"Right shit twice {a >> 2}")
print(f"Left shift {a << 1}")
print(f"Left shift twice {a << 2}")
```

```python
userId = 1005
uidb = [(userId >> 8 ) & 0xFF, userId & 0xFF]
print(uidb)
```

```python
bits = []
for b in uidb:
    for i in range(7, -1, -1):
        bits.append( (b >> i) & 1 )
print(bits)
print(np.array(bits, dtype=np.uint8))
```

```python
a = np.array([1, 0, 1, 0]*4)
print(a)
out = np.correlate(a, [1, 0, 1, 0])
print(out)
peaks, _= find_peaks(out)
print(peaks)
print(out[peaks])
```

```python
pathLoss = 0.01
chCoeff = np.sqrt(pathLoss) * ( np.sqrt(1/2) * ( np.random.randn(10000) + 1j*np.random.randn(10000) ) )
print(f"Mean Channel Power: {np.mean(np.abs(chCoeff)**2)}")
```
